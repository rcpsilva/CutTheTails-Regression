```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("./tcc-targa.csv")
df.head()
```

```python
sns.set(rc={'figure.figsize':(17,8.27)})
fig = sns.barplot(data=df, x="data_set", y="MAPE", hue="strategy")
fig.set_xticklabels(labels = df['data_set'].unique(),rotation=30) 
```

# MAPE de uma base especfica

```python
x = 'model'
dff = df[df['data_set']=='Blueberry Yield']
sns.set(rc={'figure.figsize':(17,8.27)})
fig = sns.barplot(data=dff, x=x, y="MAPE", hue="strategy")
fig.set_xticklabels(labels = df[x].unique(),rotation=30) 
```

# MAPE Geral por estratgia

```python
x = 'model'
sns.set(rc={'figure.figsize':(17,8.27)})
fig = sns.barplot(data=df, x=x, y="MAPE", hue="strategy", ci=None)
fig.set_xticklabels(labels = df[x].unique(),rotation=30) 
```
